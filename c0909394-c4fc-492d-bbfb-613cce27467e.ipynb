## Описание проекта

Вы работаете в IT-компании, которая выполняет на заказ проекты по машинному обучению. К вам обратился фермер, владелец молочного хозяйства «Вольный луг». Он хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ «ЭкоФерма».
Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже. А продавцы и технологи так и норовят приукрасить своих коровок!
Поэтому он просит вас разработать модель машинного обучения, которая поможет ему управлять рисками и принимать объективное решение о покупке. «ЭкоФерма» готова предоставить подробные данные о своих коровах. Вам нужно создать две прогнозные модели для отбора бурёнок в поголовье:
1.	Первая будет прогнозировать возможный удой коровы (целевой признак Удой);
2.	Вторая — рассчитывать вероятность получить вкусное молоко от коровы (целевой признак Вкус молока).
С помощью модели нужно отобрать коров по двум критериям:
•	средний удой за год — не менее 6000 килограммов;
•	молоко должно быть вкусным.

## Описание данных


Вы будете работать с тремя датасетами:`ferma_main.csv`, `ferma_dad.csv`, `cow_buy.csv`.

1.	Файл `ferma_main.csv` содержит данные о стаде фермера на текущий момент. Описание данных: \
	`id` — уникальный идентификатор коровы.\
	`Удой, кг` — масса молока, которую корова даёт в год (в килограммах).\
	`ЭКЕ (Энергетическая кормовая единица)` — показатель питательности корма коровы.\
	`Сырой протеин, г` — содержание сырого протеина в корме (в граммах).\
	`СПО (Сахаро-протеиновое соотношение)` — отношение сахара к протеину в корме коровы.\
	`Порода` — порода коровы.\
	`Тип пастбища` — ландшафт лугов, на которых паслась корова.\
	`порода папы_быка` — порода папы коровы.\
	`Жирность,%` — содержание жиров в молоке (в процентах).\
	`Белок,%` — содержание белков в молоке (в процентах).\
	`Вкус молока` — оценка вкуса по личным критериям фермера, бинарный признак (вкусно, не вкусно).\
	`Возраст` — возраст коровы, бинарный признак (менее_2_лет, более_2_лет).
    
Признаки в датасете `ferma_main.csv` можно разделить на группы.\
	Характеристики коровы: `id`, `Порода`, `порода папы_быка`, `Возраст`.\
	Характеристики корма: `ЭКЕ (Энергетическая кормовая единица)`, `Сырой протеин, г`, `СПО (Сахаро-протеиновое соотношение)`.\
	Характеристика пастбища: `Тип пастбища`.\
	Характеристики молока: `Удой, кг`, `Жирность,%`, `Белок,%`, `Вкус молока`.\
Такое разделение поможет вам понять, как взаимосвязаны признаки в датасете.

2.	Файл `ferma_dad.csv` хранит имя папы каждой коровы в стаде фермера. Описание данных: \
	`id` — уникальный идентификатор коровы.\
	`Имя Папы` — имя папы коровы.\
    
3.	Файл `cow_buy.csv` — это данные о коровах «ЭкоФермы», которых фермер хочет изучить перед покупкой. Описание данных: \
	`Порода` — порода коровы.\
	`Тип пастбища` — ландшафт лугов, на которых паслась корова.\
	`порода папы_быка` — порода папы коровы.\
	`Имя_папы` — имя папы коровы.\
	`Текущая_жирность,%` — содержание жиров в молоке (в процентах).\
	`Текущий_уровень_белок,%` — содержание белков в молоке (в процентах).\
	`Возраст` — возраст коровы, бинарный признак (менее_2_лет, более_2_лет).
    
Данные в файлах `ferma_main.csv` и `cow_buy.csv` похожи, но у датасета `cow_buy.csv` несколько особенностей.\
	Содержание белков и жиров в молоке указано на момент продажи — когда «ЭкоФерма» давала коровам свой корм.\
	Параметры корма `ЭКЕ (Энергетическая кормовая единица)`, `Сырой протеин, г` и `СПО (Сахаро-протеиновое соотношение)` отсутствуют. Технологи заказчика пересмотрели подход к кормлению: для новых коров планируется увеличить значения каждого из этих параметров на 5%.\
	Кроме того, отсутствуют признаки `Удой, кг` и `Вкус молока`. Это целевые признаки, именно их предстоит предсказать!

В некоторых столбцах с плавающей точкой наблюдается плавающая запятая - напишем функцию, которая будет заменять запятую на подходящий символ. Это нужно для того, чтобы при помощи `astype` привести данные к типу `float`. Если этого не сделать, интерпретатор будет ругаться на невозможность перевести строковый тип в число с плавающей точкой.

```python
def comma_to_dot(string):
    return string.replace(',', '.')
```

### Обучение второй модели линейной регрессии

Устраним нелинейность между СПО и Удоем путем преобразования признака СПО в категориальный. Граница, разделяющая значения на два кластера лежит где-то около 0.91.

Напишем функцию:

```python
def spr_to_cat(spo_value):
    if spo_value > 0.91:
        return 1
    else:
        return 0
```

```python
data_farm['спо (сахаро-протеиновое соотношение)'] = (data_farm['спо (сахаро-протеиновое соотношение)']
                                                     .apply(spr_to_cat))
```

Также известно, что для устранения нелинейной связи между Удоем и ЭКЕ нужно возвести ЭКЕ в квадрат. Для этого тоже напишем функцию:

```python
def efu_to_square(efu):
    return efu ** 2
```

```python
data_farm['эке (энергетическая кормовая единица)'] = (
    data_farm['эке (энергетическая кормовая единица)'].apply(efu_to_square))
```

Теперь, когда данные признаки изменены, можно обучить вторую модель линейной регрессии:

```python
X = data_farm.drop('удой, кг', axis=1)
y = data_farm['удой, кг']

X_train, X_test, y_train, y_test_2 = train_test_split(X, y, random_state=RANDOM_STATE)


cat_col_names = ['порода', 'тип пастбища', 'порода папы', 'возраст', 'спо (сахаро-протеиновое соотношение)']
num_col_names = ['эке (энергетическая кормовая единица)',
                 'сырой протеин, г']

encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
X_train_ohe_model_2 = encoder.fit_transform(X_train[cat_col_names])
X_test_ohe_model_2 = encoder.transform(X_test[cat_col_names])

encoder_col_names = encoder.get_feature_names_out()

scaler = StandardScaler()
X_train_scaled_model_2 = scaler.fit_transform(X_train[num_col_names])
X_test_scaled_model_2 = scaler.transform(X_test[num_col_names])

X_train_ohe_model_2 = pd.DataFrame(X_train_ohe_model_2.toarray(), columns=encoder_col_names)
X_test_ohe_model_2 = pd.DataFrame(X_test_ohe_model_2.toarray(), columns=encoder_col_names)

X_train_scaled_model_2 = pd.DataFrame(X_train_scaled_model_2, columns=num_col_names)
X_test_scaled_model_2 = pd.DataFrame(X_test_scaled_model_2, columns=num_col_names)

X_train_model_2 = pd.concat([X_train_ohe_model_2, X_train_scaled_model_2], axis=1)
X_test_model_2 = pd.concat([X_test_ohe_model_2, X_test_scaled_model_2], axis=1)

model_2_lr  = LinearRegression()
model_2_lr.fit(X_train_model_2, y_train)
predictions_model_2 = model_2_lr.predict(X_test_model_2)
```

```python
model_2_r2_score = r2_score(predictions_model_2, y_test_2)
model_2_r2_score
```

76% - не самое лучшее значение, но уже лучше, чем предыдущие 70.

### Анализ модели

Ошибки первого рода (false positive) являются более критичными, чем ошибки второго рода (false negative), при прогнозировании вкуса молока, когда главным критерием для заказчика является наличие вкусного молока. Ошибки первого рода (false positive) - это случаи, когда модель предсказывает, что молоко вкусное, но на самом деле оно невкусное. Такие ошибки в данном случае более критичны, так как если модель ошибочно предскажет, что молоко вкусное, когда оно на самом деле невкусное, потребитель может купить это молоко, ожидая, что оно вкусное. Это может негативно повлиять на удовлетворенность потребителя и доверие к модели.

Поэтому, в данном случае, стоит больше опираться на метрику precision, которая показывает, какая доля предсказанного модель вкусного молока на самом деле является вкусным. Высокая precision означает, что модель редко ошибается, предсказывая невкусное молоко как вкусное. Это важнее, чем высокий recall (полнота), так как главная цель - не предсказывать вкусным невкусное молоко.

```python
y_pred_prob = clf.predict_proba(X_test_log)[:, 1]

# Перебираем пороги от 0.2 до 0.9 с шагом 0.01
thresholds = np.linspace(0.2, 0.9, 71)

# Инициализируем списки для хранения метрик
precisions = []
recalls = []
accuracies = []

for threshold in thresholds:
    y_pred = (y_pred_prob > threshold).astype(int)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    
    precisions.append(precision)
    recalls.append(recall)
    accuracies.append(accuracy)

# Находим порог, при котором ошибки 1го рода равны 0
zero_error_threshold = thresholds[np.where(np.array(precisions) == 1)[0][0]]
print(f"Порог, при котором ошибки 1го рода равны 0: {zero_error_threshold:.2f}")

# Вычисляем метрики при этом пороге
y_pred = (y_pred_prob > zero_error_threshold).astype(int)
new_accuracy = accuracy_score(y_test, y_pred)
new_recall = recall_score(y_test, y_pred)
new_precision = precision_score(y_test, y_pred)

print(f"Новые метрики:")
print(f"Accuracy: {new_accuracy:.4f}")
print(f"Recall: {new_recall:.4f}")
print(f"Precision: {new_precision:.4f}")
```

Исходя из анализа модели логистической регрессии для прогнозирования вкуса молока, можно сделать следующие выводы и наметить дальнейшие шаги:

1) Оптимизация модели для минимизации критичных ошибок первого рода (false positive)

2) Изменение порога принадлежности к классам с 0.5 до 0.76 позволило увеличить precision до 1, что означает правильное предсказание 100% случаев вкусного молока

o	Какие метрики моделей классификации, в том числе логистической регрессии, важны при анализе рисков или экономических расчётов?

Матрица ошибок (Confusion Matrix): Показывает количество верно и неверно классифицированных объектов для каждого класса. Это базовая метрика, которая позволяет оценить общую точность модели, а также ее ошибки первого и второго рода.

Точность (Precision): Доля верно классифицированных положительных объектов среди всех объектов, предсказанных как положительные. Важна, когда последствия ложноположительных ошибок высоки.

Полнота (Recall): Доля верно классифицированных положительных объектов среди всех фактически положительных объектов. Важна, когда последствия ложноотрицательных ошибок высоки.

F1-мера (F1-score): Гармоническое среднее между точностью и полнотой. Является сбалансированной метрикой, когда важны оба типа ошибок.

Специфичность (Specificity): Доля верно классифицированных отрицательных объектов среди всех фактически отрицательных объектов. Важна, когда последствия ложноположительных ошибок высоки.

Выбор конкретных метрик зависит от бизнес-задачи и последствий ошибок классификации. Для анализа рисков и экономических расчетов наиболее важны точность, полнота и F1-мера.
